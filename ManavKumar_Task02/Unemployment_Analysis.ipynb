### Unemployment Analysis using Python

```python
# Import required libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
unemployment_data = pd.read_csv("Unemployment.csv")
```

```python
# Display first 10 rows
print("Dataset Preview:\n")
print(unemployment_data.head(10))
```

```python
# Dataset Information
print("\nDataset Info:\n")
print(unemployment_data.info())
```

```python
# Basic statistics
print("\nStatistical Summary:\n")
print(unemployment_data.describe())
```

```python
unemployment_data.columns
```

```python
#Rename column names
unemployment_data = unemployment_data.rename(columns={' Date':'Date',' Frequency':'Frequency',
                                                      ' Estimated Unemployment Rate (%)':'Estimated Unemployment Rate (%)',
                                                     ' Estimated Employed':'Estimated Employed',
                                                     ' Estimated Labour Participation Rate (%)':'Estimated Labour Participation Rate (%)'})
```

```python
# Unemployment rate distribution

plt.figure(figsize=(8,5))
sns.histplot(unemployment_data['Estimated Unemployment Rate (%)'], bins=15)

plt.title("Distribution of Unemployment Rate")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Top states with highest unemployment

state_unemployment = unemployment_data.groupby('Region')[
    'Estimated Unemployment Rate (%)'
].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
state_unemployment.head(10).plot(kind='bar')

plt.title("Top 10 Regions by Average Unemployment Rate")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
# Monthly trend

unemployment_data[' Date'] = pd.to_datetime(unemployment_data['Date'])

monthly_rate = unemployment_data.groupby('Date')[
    'Estimated Unemployment Rate (%)'
].mean()

plt.figure(figsize=(12,6))
plt.plot(monthly_rate.index, monthly_rate.values)

plt.title("Unemployment Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Average Unemployment Rate (%)")
plt.grid(True)
plt.show()
```
